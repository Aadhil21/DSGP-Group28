Remove Outliers

```python
import pandas as pd
import numpy as np

# Function to remove outliers based on IQR method
def remove_outliers(df):
    cleaned_df = df.copy()
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - 1.5 * iqr
            upper_bound = q3 + 1.5 * iqr
            cleaned_df = cleaned_df[(cleaned_df[col] >= lower_bound) & (cleaned_df[col] <= upper_bound)]
    return cleaned_df

# Check outliers
outliers_removed_df = remove_outliers(df)

# Display the shape of the original and cleaned DataFrame
print("Original DataFrame shape:", df.shape)
print("DataFrame shape after removing outliers:", outliers_removed_df.shape)

# Display the cleaned DataFrame
print(outliers_removed_df)
```

Balance Dataset

```python
from imblearn.over_sampling import SMOTE

# Separate features (X) and labels (y)
X = outliers_removed_df.drop(['Label'], axis=1)
y = outliers_removed_df['Label']

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to the dataset
X_balanced, y_balanced = smote.fit_resample(X, y)

# Create a balanced DataFrame
balanced_df = pd.concat([pd.DataFrame(X_balanced, columns=X.columns), pd.Series(y_balanced, name='Label')], axis=1)

# Display the shape of the balanced DataFrame
print("Balanced DataFrame shape:", balanced_df.shape)

# Display the class distribution in the balanced DataFrame
print("Class distribution in balanced DataFrame:")
print(balanced_df['Label'].value_counts())
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Separate features (X) and labels (y)
X = balanced_df.drop(['Label'], axis=1)
y = balanced_df['Label']

# Initialize StandardScaler
scaler = StandardScaler()

# Fit and transform the features, excluding the 'Label' column
X_scaled = scaler.fit_transform(X)

# Create a DataFrame with scaled features and the 'Label' column
scaled_df = pd.concat([pd.DataFrame(X_scaled, columns=X.columns), pd.Series(y, name='Label')], axis=1)

# Display the first few rows of the scaled DataFrame
print("Scaled DataFrame:")
print(scaled_df.head())
```
